```python
def show( text : str or float ) -> None:
    
    """
    This function uses Displacy from Spacy to display the found addresses.
    
    PARAMETERS
    ----------
        text : str
             the text to be analyzed
    """
    
    if type( text ) != str:
        
        print( "Leerer Text" )
        
    else:
        
        doc = model( text )
        
        spacy.displacy.render( doc, style = "ent" )
```
